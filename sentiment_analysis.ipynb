```python
#importing sentiment analyzer
sid=SentimentIntensityAnalyzer()#english
sidspanish=sentiment_analysis.SentimentAnalysisSpanish()#spanish

#creats loop that creates a list of sentimes per article
def headline_sentiment_summarizer_avg(headlines):
    sentiment = []
    for headline in headlines:
        sentiment_score=sid.polarity_scores(headline)['compound']
        sentiment.append(sentiment_score)
        
    return sentiment
```

```python
#creating function to be able to add arrays
def add(x,y,z):
    return x+y+z
```
